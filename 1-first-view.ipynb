```python
import pandas as pd
```

```python
adj_matrices = pd.read_pickle('data/SysFlows.pkl')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
for mat in adj_matrices:
    plt.hist(mat.flatten(), log=True)
    plt.show()
```

A HUGE variance of values of edge weights -- not good.

I'm standardizing them by making all rows sum to 1.

```python
eps = 1e-10  # In case the row sum is 0

norm_adj_matrices = []
    
for mat_i, mat in enumerate(adj_matrices):
    mat = mat / (mat.sum(axis=1)[:,None] + eps)
    norm_adj_matrices.append(mat)
    plt.hist(mat.flatten(), log=True)
    plt.show()
```

Much better.
